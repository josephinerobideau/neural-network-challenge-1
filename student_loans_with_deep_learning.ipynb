# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

---

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"].values

# Display a sample of y
y[:5]
```

```python
# Define features set X by selecting all columns but credit_ranking
X = loans_df.drop(columns=["credit_ranking"]).copy()

# Review the features DataFrame
X.head()
```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_scaler = scaler.fit(X_train)

# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

---

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
number_inputs = len(X_train.columns)

# Review the number of features
number_inputs
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer_1 = 6

# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer_2 = 3

# Define the number of neurons in the output layer
hidden_nodes_layer_3 = 1
```

```python
# Create the Sequential model instance
model = Sequential()

# Add the first hidden layer
model.add(Dense(units=hidden_nodes_layer_1, input_dim=number_inputs, activation="relu"))

# Add the second hidden layer
model.add(Dense(units=hidden_nodes_layer_2, activation="relu"))

# Add the output layer to the model specifying the number of output neurons and activation function
model.add(Dense(units=hidden_nodes_layer_3, activation="sigmoid"))
```

```python
# Display the Sequential model summary
model.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
# Fit the model using 50 epochs and the training data
fit_model = model.fit(X_train_scaled, y_train, epochs=50)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test, verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Step 4: Save and export your model to a keras file, and name the file `student_loans.keras`.


```python
# Set the model's file path
file_path = Path("student_loans.keras")

# Export your model to a keras file
model.save(file_path)
```

---
## Predict Loan Repayment Success by Using your Neural Network Model

### Step 1: Reload your saved model.

```python
# Set the model's file path
file_path = Path("student_loans.keras")

# Load the model to a new object
nn_imported = tf.keras.models.load_model(file_path)
```

### Step 2: Make predictions on the testing data and save the predictions to a DataFrame.

```python
# Make predictions with the test data
predictions = nn_imported.predict(X_test_scaled, verbose = 2)

# Display a sample of the predictions
predictions[:5]
```

```python
# Save the predictions to a DataFrame and round the predictions to binary results
predictions_df = pd.DataFrame(data = predictions, columns=["predictions"])
predictions_df["predictions"] = round(predictions_df["predictions"], 0)
predictions_df.head()
```

### Step 4: Display a classification report with the y test data and predictions

```python
# Print the classification report with the y test data and predictions
print(classification_report(y_test, predictions_df["predictions"].values))
```

**3. Describe two real-world challenges that you would take into consideration while building a recommendation system for student loans. Explain why these challenges would be of concern for a student loan recommendation system.**

One real-world challenge would be the amount of data and the uniqueness of the data. This could be problematic because if the data is not significant enough for a wide array of students with varying needs, the recommendation would be biased and, therefore, be giving a very general recommendation that would not be an ideal fit for the student. For example, say the average student needs a loan for four years of university, has an 85 average in high school (relatively promising), and is majoring in Business (likely to make 60k/yr +). This person's recommendation should be very different than a student who has already done two years, has two years left, has a 3.0 GPA or less, and is a liberal arts major. Though both students could very well be capable of the same things and paying back the loan, the facts are that the second student's tuition will be less money, for less time, with potentially less promising grades for graduation and a less promising career than would be able to pay back the loan. Overall, this example shows us the importance of having well-rounded data capable of giving each student a unique recommendation that will fit their circumstances.

The second challenge could be how often the model is trained and how it is being updated. As the average student demographics change, as there are more possibilities for the average student's need and more possibilities for unique needs (but not necessarily as expected), the recommendation model needs to have regular training and updates to ensure students are getting the most up to date and accurate recommendation they can get. Otherwise, the results could be similar to those of the having biased data, and therefore, getting a recommendation that is only somewhat reliable or as unique to the student as it should be.
